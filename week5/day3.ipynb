```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1] 
plt.plot(x, y)
plt.title('Product Demand at Different Price Points')
plt.xlabel('Price')
plt.ylabel('Demand')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm
plt.scatter(x, y)
plt.title('Children\'s Heights at Different Ages')
plt.xlabel('Age (years)')
plt.ylabel('Height (cm)')
plt.show()
```
